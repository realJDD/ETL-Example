```python
import pandas as pd
```

```python
url = 'http://www.espn.com/mlb/stats/batting/_/year/2012/sort/homeRuns/order/true'
```

```python
table = pd.read_html(url)
table
```

```python
type(table)
```

```python
df = table[0]
df.dropna()
df.columns=['RK', 'PLAYER', 'TEAM','AB','R','H','2B','3B','HR','RBI','SB','CS','BB','SO','AVG','OBP','SLG','OPS','WAR']

```

```python
df = df.iloc[2:]
```

```python
df.head()
```

```python
HR_leaders = []
```

```python
for i in range(0, len(df)):
    HR_leaders.append({"Player" : df.iloc[i]['PLAYER'], "HR" : df.iloc[i]['HR']})
    print(df.iloc[i]['RK'], df.iloc[i]['PLAYER'], df.iloc[i]['HR'])
```
